```python
import pandas as pd
pd.set_option('display.precision', 3)
pd.set_option('display.max_colwidth', None)
```

```python
beam = 'b4'
bb_df_sel = bb_dfs[beam]
# elements_for_illustration = ['bb_ho.c1b1_00', 'bb_ho.c2b1_00', 'bb_ho.c5b1_00', 'bb_ho.c8b1_00',
#                             'bb_lr.l1b1_01', 'bb_lr.r1b1_02',]

elements_for_illustration = [
    'bb_lr.l1b1_25', 'bb_lr.l1b1_01',
    'bb_ho.l1b1_05', 'bb_ho.c1b1_00', 'bb_ho.l1b1_05', 
    'bb_lr.r1b1_01', 'bb_lr.r1b1_25',]

if beam in ['b2', 'b4']:
    for ii, ee in enumerate(elements_for_illustration):
        elements_for_illustration[ii] = ee.replace('b1', 'b2')

if beam in ['b3', 'b4']:
    elements_for_illustration = elements_for_illustration[::-1]
```

```python
cols_to_show = ['other_charge_ppb', 'separation_x', 'separation_y', 'dpx', 'dpy',
              'alpha', 'phi', 'other_Sigma_11', 'other_Sigma_13', 'other_Sigma_33',]
              
df_show = pd.DataFrame()
for kk in cols_to_show:
    if kk.startswith('other_'):
        new_kk = kk.replace('other_', '')
    else:
        new_kk = kk
    df_show[new_kk] = bb_df_sel[kk]
df_show.loc[elements_for_illustration, :].head(10)
```
